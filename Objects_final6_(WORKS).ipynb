```python
def board(PIECE_COUNT):   
  if PIECE_COUNT % 2 != 0:
    board = np.zeros([PIECE_COUNT,PIECE_COUNT])
    coor = int((PIECE_COUNT-1) / 2)
    board[coor,coor] = 1
    
  else:
    board = np.zeros([PIECE_COUNT + 1,PIECE_COUNT + 1])
    coor = int(PIECE_COUNT / 2)
    board[coor,coor] = 1

  return board
```

```python
def form_objects():
  objects = []
  if PIECE_COUNT % 2 != 0:
    coor = int((PIECE_COUNT-1) / 2)
  else:
    coor = int(PIECE_COUNT / 2)
  
  for i in range(PIECE_COUNT-1):
    while len(objects) < PIECE_COUNT-1:
      x,y = random.randint(0,PIECE_COUNT-1),random.randint(0,PIECE_COUNT-1)
      if (x,y) not in objects and (x,y) != (coor,coor):
        objects.append((x,y))


  return objects
```

```python
def trim_zeros(arr, margin=0):

    s = []
    for dim in range(arr.ndim):
        start = 0
        end = -1
        slice_ = [slice(None)]*arr.ndim

        go = True
        while go:
            slice_[dim] = start
            go = not np.any(arr[tuple(slice_)])
            start += 1
        start = max(start-1-margin, 0)

        go = True
        while go:
            slice_[dim] = end
            go = not np.any(arr[tuple(slice_)])
            end -= 1
        end = arr.shape[dim] + min(-1, end+1+margin) + 1

        s.append(slice(start,end))
    return arr[tuple(s)]
```

```python
# Rotation check.
final_list = []
rots = []

for element in unique_list:

  for i in range(1,4):
    a = np.rot90(element,i).tolist()
    rots.append(a)

  flipped = np.fliplr(element).tolist()

  for i in range(4):
    b = np.rot90(flipped,i).tolist()
    rots.append(b)


  if element in rots:
    # Remove all the occurences of the original element
    rots = [item for item in rots if item != element]

  rots_2 = []
  for item in rots:
    if item not in rots_2:
      rots_2.append(item)

  for vals in rots_2:
    if vals not in final_list:
      final_list.append(vals)

  for item in final_list:
    if item in unique_list:
      unique_list.remove(item)
print(f"Results count: {len(unique_list)}")
print(f'Type of unique_list elements: {type(unique_list[0])}')
```
